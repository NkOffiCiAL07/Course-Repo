Write a function “minout(l)” that takes a list of distinct natural numbers and returns
the smallest natural number not present in l. For example, if 0 is not in l, then the function should
return 0, else if 0 is in l but 1 is not in l, then it should return 1 etc.

```python
# random list
l = [3, 5, 1, 0, 5, 2, 8]

def minout(list):
    list.sort()
    mex = 0
    lst = []
    for i in range(len(list)):
        if mex == list[i]:
            mex += 1
    
    return mex

print(minout(l))
```

### Exercise-2

Write a function “overlap” that take two lists of integers and returns the number of
distinct values that occur in both lists. For instance; overlap([1,3,4,7],[9,13,7,5,1,2]) should return 2
since the numbers 1 and 7 occur in both lists.

```python
# random two lists
l1 = [1, 3, 4, 7]
l2 = [9, 13, 7, 5, 1, 2]

def overlap(list1, list2):
    
    count = 0
    for val in list1:
        if val in list2:
            count += 1
            
    return count

print(overlap(l1, l2))
    
```

### Exercise-3

Generate a list L of 100 random numbers between 1 and 300. A ‘scatter plot’ uses dots to
represent values of two different numeric variables. Use scatter() function in plt to plot the values
of L in y-axis.

```python
# importing matplotlib module 
from matplotlib import pyplot as plt

# importing numpy module
import numpy as np

y = []
for i in range(100):
    y.append(np.random.randint(1, 300))

x = range(1, 101, 1)

plt.scatter(x, y)

# Labelling X-axis and Y-axis
plt.xlabel("X-Values")
plt.ylabel("Y-Values")

# displaying the title
plt.title("Scatter Plot")
plt.show()
```

### Exercise-4

A simple plot uses lines to represent values of two different numeric variables. Use
plot() function in plt to plot the values of L in y-axis.

```python
fig = plt.figure(figsize =(10, 5))
plt.bar(x, y)

# Labelling X-axis and Y-axis
plt.xlabel("X-Values")
plt.ylabel("Y-Values")

# displaying the title
plt.title("Bar Plot")
```

### Exercise-6

The functions scatter(), bar() and plot() take several other parameters as inputs. Use
the official documentation of matplotlib available on https://matplotlib.org/ to learn more about
these parameters and change the colors of the three plots obtained in the previous exercises.

```python
plt.scatter(x, y, c = 'red')

# Labelling X-axis and Y-axis
plt.xlabel("X-Values")
plt.ylabel("Y-Values")

# displaying the title
plt.title("Scatter Plot")
plt.show()
```

```python
fig = plt.figure(figsize =(10, 5))
plt.bar(x, y, color = ['red', 'blue', 'black'])

# Labelling X-axis and Y-axis
plt.xlabel("X-Values")
plt.ylabel("Y-Values")

# displaying the title
plt.title("Bar Plot")
```
